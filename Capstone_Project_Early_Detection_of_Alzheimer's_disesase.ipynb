```python
# Run in Colab (mount Drive first if using Google Drive)

import os, math, joblib, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image

import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader, random_split, WeightedRandomSampler
from torchvision import transforms, models
from torchvision.transforms import RandAugment, RandomResizedCrop
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_recall_fscore_support

import timm, seaborn as sns, gradio as gr

# ----------------- CONFIG -----------------
BASE_DIR = "/content/drive/MyDrive/alzheimers-demo"   # << change to your Drive folder
IMG_ROOT = os.path.join(BASE_DIR, "extracted_images")
META_FILE = os.path.join(BASE_DIR, "meta-data.xlsx")
MODEL_DIR = os.path.join(BASE_DIR, "models")
ARTIFACTS_DIR = os.path.join(BASE_DIR, "artifacts")
os.makedirs(MODEL_DIR, exist_ok=True); os.makedirs(ARTIFACTS_DIR, exist_ok=True)

BATCH_SIZE = 16
NUM_EPOCHS = 24
LEARNING_RATE = 3e-5
WEIGHT_DECAY = 1e-2
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42
torch.manual_seed(SEED)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(SEED)
np.random.seed(SEED)

CLASS_NAMES = ["Healthy","MCI","Mild AD"]

IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

# ----------------- TRANSFORMS -----------------
train_transform = transforms.Compose([
    RandomResizedCrop(224, scale=(0.75, 1.0)),
    transforms.RandomHorizontalFlip(0.5),
    transforms.RandomRotation(8),
    RandAugment(num_ops=2, magnitude=7),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
])
infer_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
])

# ----------------- DATASET -----------------
class AlzheimerDataset(Dataset):
    def __init__(self, excel_file, image_root, transform=None, meta_norm=True):
        self.transform = transform
        self.image_root = image_root
        self.df = pd.read_excel(excel_file)
        self.df['ID'] = self.df['ID'].astype(str)
        # label mapping to 3 classes
        self.df['label3'] = self.df['CDR'].apply(lambda c: 0 if c==0 else 1 if c==0.5 else 2)
        self.samples = []
        for root, _, files in os.walk(image_root):
            for fname in files:
                if fname.lower().endswith(('.png','.jpg','.jpeg')):
                    img_id = fname.split("_mpr")[0]
                    row = self.df[self.df['ID'] == str(img_id)]
                    if not row.empty:
                        meta_vals = row[['Age','MMSE']].iloc[0].astype(float).values
                        lbl = int(row['label3'].iloc[0])
                        self.samples.append((os.path.join(root,fname), meta_vals, lbl))
        if len(self.samples) == 0:
            raise RuntimeError("No samples found. Check IMG_ROOT and META_FILE.")
        metas = np.stack([s[1] for s in self.samples], axis=0)
        self.meta_mean = metas.mean(axis=0)
        self.meta_std = metas.std(axis=0) + 1e-8
        self.meta_norm = meta_norm

    def __len__(self): return len(self.samples)

    def __getitem__(self, idx):
        path, meta_vals, lbl = self.samples[idx]
        img = Image.open(path).convert('RGB')
        if self.transform: img = self.transform(img)
        meta = (meta_vals - self.meta_mean) / self.meta_std if self.meta_norm else meta_vals
        return img, torch.tensor(meta, dtype=torch.float32), torch.tensor(lbl, dtype=torch.long)

# ----------------- LOADERS -----------------
def get_loaders(dataset, batch_size=BATCH_SIZE, seed=SEED, weighted_sampler=True):
    n = len(dataset)
    train_n = int(0.7 * n); val_n = int(0.15 * n); test_n = n - train_n - val_n
    train_ds, val_ds, test_ds = random_split(dataset, [train_n, val_n, test_n], generator=torch.Generator().manual_seed(seed))
    train_targets = [dataset.samples[i][2] for i in train_ds.indices]
    class_counts = Counter(train_targets)
    sample_weights = [1.0 / class_counts[t] for t in train_targets]
    if weighted_sampler:
        sampler = WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)
        train_loader = DataLoader(train_ds, batch_size=batch_size, sampler=sampler, pin_memory=torch.cuda.is_available())
    else:
        train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, pin_memory=torch.cuda.is_available())
    val_loader = DataLoader(val_ds, batch_size=batch_size, pin_memory=torch.cuda.is_available())
    test_loader = DataLoader(test_ds, batch_size=batch_size, pin_memory=torch.cuda.is_available())
    return train_loader, val_loader, test_loader, train_ds, val_ds, test_ds

# ----------------- METRICS UTIL -----------------
def compute_metrics(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)
    return {"accuracy": acc, "precision": prec, "recall": rec, "f1": f1}

# ----------------- MIXUP & CRITERION -----------------
def mixup_data(x, y, meta=None, alpha=0.2):
    if alpha <= 0:
        return x, meta, y, None, 1.0
    lam = np.random.beta(alpha, alpha)
    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    if meta is not None:
        mixed_meta = lam * meta + (1 - lam) * meta[index]
    else:
        mixed_meta = None
    return mixed_x, mixed_meta, y_a, y_b, lam

def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

def get_criterion(class_weights=None, label_smoothing=0.0):
    if class_weights is not None:
        weights_t = torch.tensor(class_weights, dtype=torch.float32).to(DEVICE)
    else:
        weights_t = None
    try:
        return nn.CrossEntropyLoss(weight=weights_t, label_smoothing=label_smoothing)
    except TypeError:
        return nn.CrossEntropyLoss(weight=weights_t)

# ----------------- MODEL FACTORIES -----------------
def create_vit_only(num_classes=3, pretrained=True, in21k=False):
    model_name = "vit_base_patch16_224_in21k" if in21k else "vit_base_patch16_224"
    model = timm.create_model(model_name, pretrained=pretrained, num_classes=num_classes)
    try:
        if hasattr(model, "set_grad_checkpointing"):
            model.set_grad_checkpointing(True)
    except Exception:
        pass
    return model

class ViTMetaFiLM(nn.Module):
    """
    Stable ViT + FiLM meta fusion:
    - gamma/beta zero-initialized
    - tanh scaled by fm_scale to keep modulation small initially
    - optional concat of small meta projection to classifier
    """
    def __init__(self, metadata_dim=2, num_classes=3, pretrained=True, in21k=False,
                 meta_hidden=64, fm_scale=0.07, use_concat=True):
        super().__init__()
        model_name = "vit_base_patch16_224_in21k" if in21k else "vit_base_patch16_224"
        self.vit = timm.create_model(model_name, pretrained=pretrained, num_classes=0)
        if hasattr(self.vit, 'head'):
            try: self.vit.head = nn.Identity()
            except Exception: pass

        D = getattr(self.vit, 'num_features', None) or getattr(self.vit, 'embed_dim', 768)

        # small meta network
        self.meta_net = nn.Sequential(
            nn.Linear(metadata_dim, meta_hidden),
            nn.ReLU(),
            nn.LayerNorm(meta_hidden),
            nn.Linear(meta_hidden, meta_hidden),
            nn.ReLU(),
            nn.LayerNorm(meta_hidden),
        )

        self.gamma = nn.Linear(meta_hidden, D)
        self.beta  = nn.Linear(meta_hidden, D)

        # meta projection for concat path
        self.meta_proj = nn.Linear(meta_hidden, min(64, meta_hidden))
        self.use_concat = use_concat

        if self.use_concat:
            self.classifier = nn.Sequential(
                nn.Linear(D + self.meta_proj.out_features, 512),
                nn.ReLU(),
                nn.Dropout(0.4),
                nn.Linear(512, num_classes)
            )
        else:
            self.classifier = nn.Sequential(
                nn.Linear(D, 512),
                nn.ReLU(),
                nn.Dropout(0.4),
                nn.Linear(512, num_classes)
            )

        # small scaling factor and zero-init FiLM heads
        self.fm_scale = float(fm_scale)
        nn.init.zeros_(self.gamma.weight); nn.init.zeros_(self.gamma.bias)
        nn.init.zeros_(self.beta.weight);  nn.init.zeros_(self.beta.bias)

        try:
            if hasattr(self.vit, "set_grad_checkpointing"):
                self.vit.set_grad_checkpointing(True)
        except Exception:
            pass

    def forward(self, img, meta):
        # extract CLS token feature (matching timm impl)
        if hasattr(self.vit, 'patch_embed') and hasattr(self.vit, 'blocks') and hasattr(self.vit, 'pos_embed'):
            x = self.vit.patch_embed(img)
            cls_token = self.vit.cls_token.expand(img.shape[0], -1, -1)
            x = torch.cat((cls_token, x), dim=1)
            x = x + self.vit.pos_embed
            x = self.vit.pos_drop(x)
            for blk in self.vit.blocks:
                x = blk(x)
            x = self.vit.norm(x)
            img_feat = x[:, 0, :]
        else:
            if hasattr(self.vit, 'forward_features'):
                feat = self.vit.forward_features(img)
            else:
                feat = self.vit(img)
            if feat.ndim == 2:
                img_feat = feat
            elif feat.ndim == 3:
                img_feat = feat[:,0,:]
            else:
                img_feat = feat.view(feat.size(0), -1)

        meta_h = self.meta_net(meta)
        gamma = self.gamma(meta_h)
        beta  = self.beta(meta_h)

        # stable squashing + small scale
        gamma = torch.tanh(gamma) * self.fm_scale
        beta  = torch.tanh(beta)  * self.fm_scale

        mod_feat = img_feat * (1.0 + gamma) + beta

        if self.use_concat:
            meta_p = self.meta_proj(meta_h)
            combined = torch.cat([mod_feat, meta_p], dim=1)
            out = self.classifier(combined)
        else:
            out = self.classifier(mod_feat)

        return out

def create_vit_meta(num_classes=3, pretrained=True, in21k=False):
    # slightly conservative fm_scale and use concat path
    return ViTMetaFiLM(metadata_dim=2, num_classes=num_classes, pretrained=pretrained, in21k=in21k,
                       meta_hidden=64, fm_scale=0.07, use_concat=True)

def create_resnet_meta(num_classes=3, pretrained=True):
    weights = models.ResNet18_Weights.DEFAULT if pretrained else None
    backbone = models.resnet18(weights=weights)
    backbone.fc = nn.Identity()
    class ResNetMeta(nn.Module):
        def __init__(self, backbone, metadata_dim=2, num_classes=3):
            super().__init__()
            self.backbone = backbone
            self.meta_fc = nn.Sequential(nn.Linear(metadata_dim,64), nn.ReLU(), nn.LayerNorm(64))
            self.fuse = nn.Sequential(nn.Linear(512 + 64, 256), nn.ReLU(), nn.Dropout(0.3))
            self.classifier = nn.Linear(256, num_classes)
        def forward(self, img, meta):
            img_feat = self.backbone(img)
            meta_feat = self.meta_fc(meta)
            x = torch.cat([img_feat, meta_feat], dim=1)
            x = self.fuse(x)
            return self.classifier(x)
    return ResNetMeta(backbone, metadata_dim=2, num_classes=num_classes)

def create_cnn_transformer(num_classes=3, pretrained=True):
    weights = models.ResNet18_Weights.DEFAULT if pretrained else None
    resnet = models.resnet18(weights=weights)
    cnn = nn.Sequential(*list(resnet.children())[:-2])
    class SmallTransformerHybrid(nn.Module):
        def __init__(self, cnn, embed_dim=512, num_heads=8, mlp_dim=2048, num_layers=1, num_classes=3):
            super().__init__()
            self.cnn = cnn
            self.flatten = nn.Flatten(2)
            self.transpose = lambda x: x.permute(0,2,1)
            encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads, dim_feedforward=mlp_dim, batch_first=True)
            self.transformer = nn.TransformerEncoder(encoder_layer,num_layers=num_layers)
            self.head = nn.Sequential(nn.LayerNorm(embed_dim), nn.Linear(embed_dim,num_classes))
        def forward(self,x):
            x = self.cnn(x)
            x = self.flatten(x)
            x = self.transpose(x)
            x = self.transformer(x)
            x = x.mean(dim=1)
            return self.head(x)
    return SmallTransformerHybrid(cnn, embed_dim=512, num_heads=8, mlp_dim=2048, num_layers=1, num_classes=num_classes)

# ----------------- OPTIMIZER & SCHEDULER -----------------
def get_optimizer_and_scheduler(model, lr, weight_decay, num_epochs, warmup_epochs=3, head_lr_factor=5.0):
    head_keys = ('classifier', 'meta', 'meta_net', 'gamma', 'beta', 'meta_fc', 'fuse', 'head', 'meta_proj')
    head_params = []
    backbone_params = []
    for name, p in model.named_parameters():
        if any(k in name for k in head_keys):
            head_params.append(p)
        else:
            backbone_params.append(p)
    param_groups = []
    if head_params:
        param_groups.append({'params': head_params, 'lr': lr * head_lr_factor})
    if backbone_params:
        param_groups.append({'params': backbone_params, 'lr': lr})
    if not param_groups:
        optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    else:
        optimizer = torch.optim.AdamW(param_groups, lr=lr, weight_decay=weight_decay)
    def lr_lambda(epoch):
        if epoch < warmup_epochs:
            return float(epoch + 1) / float(max(1, warmup_epochs))
        else:
            t = (epoch - warmup_epochs) / max(1, (num_epochs - warmup_epochs))
            return 0.5 * (1.0 + math.cos(math.pi * t))
    scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)
    return optimizer, scheduler

# ----------------- TRAIN / EVAL -----------------
def train_model(model, train_loader, val_loader, uses_meta=False, num_epochs=NUM_EPOCHS,
                lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY, mixup_alpha=0.05, label_smoothing=0.0,
                freeze_backbone_epochs=0, class_weights=None, head_lr_factor=5.0):
    model.to(DEVICE)
    criterion = get_criterion(class_weights, label_smoothing=label_smoothing)
    optimizer, scheduler = get_optimizer_and_scheduler(model, lr, weight_decay, num_epochs, warmup_epochs=3, head_lr_factor=head_lr_factor)
    history = {'train_loss':[], 'val_acc':[]}

    # freeze backbone initially except head params
    for name, p in model.named_parameters():
        if ('vit' in name or 'backbone' in name) and not any(h in name for h in ['classifier','meta','meta_net','gamma','beta','meta_fc','fuse','meta_proj']):
            p.requires_grad = False

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        if epoch == freeze_backbone_epochs:
            for name, p in model.named_parameters():
                p.requires_grad = True
            print(f"[{epoch+1}] Unfroze backbone (fine-tuning entire model)")

        for imgs, metas, labels in train_loader:
            imgs, metas, labels = imgs.to(DEVICE), metas.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()

            # mixup (images + meta together)
            if mixup_alpha > 0:
                imgs_m, metas_m, y_a, y_b, lam = mixup_data(imgs, labels, meta=metas, alpha=mixup_alpha)
                out = model(imgs_m, metas_m) if uses_meta else model(imgs_m)
                loss = mixup_criterion(criterion, out, y_a, y_b, lam)
            else:
                out = model(imgs, metas) if uses_meta else model(imgs)
                loss = criterion(out, labels)

            # diagnostics: print FiLM stats occasionally
            if hasattr(model, 'gamma') and epoch % 4 == 0:
                try:
                    with torch.no_grad():
                        meta_h_sample = model.meta_net(metas)
                        g_mean = model.gamma(meta_h_sample).mean().item()
                        b_mean = model.beta(meta_h_sample).mean().item()
                    print(f"  [diag] gamma_mean={g_mean:.6f} beta_mean={b_mean:.6f}")
                except Exception:
                    pass

            loss.backward()

            # gradient norm diagnostics (head vs backbone)
            total_norm_sq = 0.0
            head_norm_sq = 0.0
            for name_p, p in model.named_parameters():
                if p.grad is None: continue
                gnorm = p.grad.detach().data.norm(2).item()
                total_norm_sq += gnorm ** 2
                if any(k in name_p for k in ('classifier','meta','gamma','beta','meta_net','meta_fc','fuse','meta_proj')):
                    head_norm_sq += gnorm ** 2
            total_norm = total_norm_sq ** 0.5
            head_norm = head_norm_sq ** 0.5
            # print small diagnostic every N iterations or rarely to avoid clutter
            # (we print only at the first batch of epochs 0/4/8... to keep logs readable)
            # print(f"  [diag] grad_norm total={total_norm:.4f} head={head_norm:.4f}")

            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            running_loss += loss.item() * imgs.size(0)

        scheduler.step()
        avg_loss = running_loss / len(train_loader.dataset)

        # validation
        model.eval()
        y_true=[]; y_pred=[]
        with torch.no_grad():
            for imgs, metas, labels in val_loader:
                imgs, metas, labels = imgs.to(DEVICE), metas.to(DEVICE), labels.to(DEVICE)
                outputs = model(imgs, metas) if uses_meta else model(imgs)
                if isinstance(outputs, (tuple, list)):
                    outputs = outputs[0]
                preds = outputs.argmax(1).cpu().numpy()
                y_pred.extend(preds); y_true.extend(labels.cpu().numpy())
        val_acc = accuracy_score(y_true, y_pred)
        history['train_loss'].append(avg_loss); history['val_acc'].append(val_acc)
        print(f"[{epoch+1}/{num_epochs}] loss={avg_loss:.4f} val_acc={val_acc:.4f} lr_head={optimizer.param_groups[0]['lr']:.2e}")

    return model, history

def evaluate_model(model, loader, uses_meta=False):
    model.eval(); y_true=[]; y_pred=[]
    with torch.no_grad():
        for imgs, metas, labels in loader:
            imgs, metas, labels = imgs.to(DEVICE), metas.to(DEVICE), labels.to(DEVICE)
            outputs = model(imgs, metas) if uses_meta else model(imgs)
            if isinstance(outputs, (tuple, list)): outputs = outputs[0]
            preds = outputs.argmax(1).cpu().numpy()
            y_pred.extend(preds); y_true.extend(labels.cpu().numpy())
    cm = confusion_matrix(y_true, y_pred)
    report = classification_report(y_true, y_pred, output_dict=True, zero_division=0)
    metrics = compute_metrics(y_true, y_pred)
    return metrics, report, cm, y_true, y_pred

# ----------------- MAIN PIPELINE -----------------
dataset = AlzheimerDataset(META_FILE, IMG_ROOT, transform=train_transform, meta_norm=True)
train_loader, val_loader, test_loader, train_ds, val_ds, test_ds = get_loaders(dataset, batch_size=BATCH_SIZE)

# compute class weights from training split
train_targets = [dataset.samples[i][2] for i in train_ds.indices]
counts = np.array([train_targets.count(i) for i in range(len(CLASS_NAMES))], dtype=np.float32)
class_weights = (counts.sum() / (counts + 1e-8))
class_weights = class_weights / class_weights.sum()
print("Class weights (normalized):", class_weights)

# save meta stats
np.savez(os.path.join(MODEL_DIR, "meta_stats.npz"), mean=dataset.meta_mean, std=dataset.meta_std)
print("Saved meta normalization stats.")

# per-model training config (biased toward ViT_meta stability)
models_to_run = {
    "ViT_only": {
        "factory": (lambda: create_vit_only(num_classes=3, pretrained=True, in21k=True)),
        "uses_meta": False, "fname":"vit_only_final.pth",
        "epochs": 20, "lr": 3e-5,  "mixup":0.10, "freeze_epochs":2, "head_lr_factor":1.0, "label_smoothing":0.0
    },
    "ViT_meta": {
        "factory": (lambda: create_vit_meta(num_classes=3, pretrained=True, in21k=True)),
        "uses_meta": True,  "fname":"vit_meta_final.pth",
        "epochs": 28, "lr": 3e-5,  "mixup":0.05, "freeze_epochs":0, "head_lr_factor":5.0, "label_smoothing":0.02
    },
    "ResNet18_meta": {
        "factory": (lambda: create_resnet_meta(num_classes=3, pretrained=True)),
        "uses_meta": True,  "fname":"resnet18_meta_final.pth",
        "epochs": 20, "lr": 1e-4,  "mixup":0.10, "freeze_epochs":1, "head_lr_factor":4.0, "label_smoothing":0.02
    },
    "CNN_Transformer": {
        "factory": (lambda: create_cnn_transformer(num_classes=3, pretrained=True)),
        "uses_meta": False, "fname":"cnn_transformer_final.pth",
        "epochs": 20, "lr": 1e-4,  "mixup":0.10, "freeze_epochs":1, "head_lr_factor":1.0, "label_smoothing":0.0
    },
}

all_results = {}
for name, cfg in models_to_run.items():
    print(f"\n=== Training {name} ===")
    model = cfg["factory"]()
    try:
        if hasattr(model, "set_grad_checkpointing"):
            model.set_grad_checkpointing(True)
    except Exception:
        pass

    trained_model, history = train_model(
        model,
        train_loader, val_loader,
        uses_meta=cfg["uses_meta"],
        num_epochs=cfg["epochs"],
        lr=cfg["lr"],
        weight_decay=WEIGHT_DECAY,
        mixup_alpha=cfg["mixup"],
        label_smoothing=cfg["label_smoothing"],
        freeze_backbone_epochs=cfg["freeze_epochs"],
        class_weights=class_weights,
        head_lr_factor=cfg["head_lr_factor"]
    )

    metrics, report, cm, y_true, y_pred = evaluate_model(trained_model, test_loader, uses_meta=cfg["uses_meta"])
    model_path = os.path.join(MODEL_DIR, cfg["fname"])
    try:
        torch.save(trained_model.state_dict(), model_path)
    except Exception:
        torch.save(trained_model, model_path)
    joblib.dump(report, os.path.join(ARTIFACTS_DIR, f"{name}_report.pkl"))
    np.save(os.path.join(ARTIFACTS_DIR, f"{name}_confusion.npy"), cm)
    plt.figure(); plt.plot(history['val_acc']); plt.title(f"{name} val_acc"); plt.xlabel("epoch"); plt.ylabel("val_acc")
    plt.savefig(os.path.join(ARTIFACTS_DIR, f"{name}_valacc.png")); plt.close()
    all_results[name] = {"metrics":metrics, "cm":cm, "history":history, "weights":model_path}
    print(f"Saved {name} artifacts to {model_path}")
    print(f"{name} metrics:", metrics)

# summary CSV
summary_rows=[]
for n,r in all_results.items():
    m=r['metrics']; summary_rows.append({"model":n, **m})
summary_df = pd.DataFrame(summary_rows)
summary_df.to_csv(os.path.join(ARTIFACTS_DIR,"models_comparison_summary.csv"), index=False)
print("\n=== Final summary ===")
print(summary_df)

# ----------------- GRADIO INFERENCE -----------------
infer_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
])

meta_stats_path = os.path.join(MODEL_DIR, "meta_stats.npz")
if os.path.exists(meta_stats_path):
    npz = np.load(meta_stats_path); meta_mean = npz['mean']; meta_std = npz['std']
else:
    meta_mean = np.array([70.0, 28.0]); meta_std = np.array([10.0, 2.0])
print("meta_mean:", meta_mean, "meta_std:", meta_std)

model_files = {
    "ViT_only": ("vit_only_final.pth", False, lambda: create_vit_only(num_classes=3, pretrained=False, in21k=True)),
    "ViT_meta": ("vit_meta_final.pth", True,  lambda: create_vit_meta(num_classes=3, pretrained=False, in21k=True)),
    "ResNet18_meta": ("resnet18_meta_final.pth", True, lambda: create_resnet_meta(num_classes=3, pretrained=False)),
    "CNN_Transformer": ("cnn_transformer_final.pth", False, lambda: create_cnn_transformer(num_classes=3, pretrained=False)),
}

loaded_models = {}
for display_name, (fname, uses_meta, factory) in model_files.items():
    p = os.path.join(MODEL_DIR, fname)
    if not os.path.exists(p):
        print(f"WARNING: missing model file: {p} (skipping {display_name})"); continue
    print("Loading", display_name, "->", p)
    model = factory()
    sd = torch.load(p, map_location=DEVICE)
    loaded_ok = False
    if isinstance(sd, dict):
        if any(k.startswith("module.") for k in sd.keys()):
            sd = {k.replace("module.", ""): v for k, v in sd.items()}
        try:
            model.load_state_dict(sd); loaded_ok=True
        except Exception as e:
            try:
                model.load_state_dict(sd, strict=False); loaded_ok=True; print(f"Loaded {display_name} with strict=False.")
            except Exception as e2:
                print(f"Failed to load state_dict for {display_name}: {e2}"); loaded_ok=False
    if not loaded_ok:
        if isinstance(sd, nn.Module):
            model = sd; print(f"Loaded full model object for {display_name}.")
        else:
            print(f"Couldn't load weights for {display_name}. Skipping."); continue
    model.to(DEVICE).eval()
    loaded_models[display_name] = {"model": model, "uses_meta": uses_meta}

confusion_mats = {}
for name in model_files:
    p = os.path.join(ARTIFACTS_DIR, f"{name}_confusion.npy")
    confusion_mats[name] = np.load(p) if os.path.exists(p) else None

def tta_predict_probs(model, pil_image, meta_tensor, uses_meta=False, tta_transforms=4):
    probs = None
    for t in range(tta_transforms):
        img_aug = pil_image.copy()
        if t % 2 == 1: img_aug = transforms.functional.hflip(img_aug)
        if t % 4 == 2: img_aug = transforms.functional.rotate(img_aug, angle=5)
        img_tensor = infer_transform(img_aug).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            try:
                out = model(img_tensor, meta_tensor) if uses_meta else model(img_tensor)
            except TypeError:
                out = model(img_tensor)
            if isinstance(out, (tuple, list)): out = out[0]
            outp = torch.softmax(out, dim=1).cpu().numpy()[0]
        probs = outp if probs is None else probs + outp
    return probs / tta_transforms

def predict_all(image, age, mmse):
    if image is None: return pd.DataFrame(), None, None
    img_pil = image.convert("RGB")
    meta = np.array([age, mmse], dtype=float)
    meta_norm = (meta - meta_mean) / meta_std
    meta_tensor = torch.tensor(meta_norm, dtype=torch.float32).unsqueeze(0).to(DEVICE)

    rows=[]
    for name, info in loaded_models.items():
        model = info["model"]; uses_meta = info["uses_meta"]
        try:
            probs = tta_predict_probs(model, img_pil, meta_tensor if uses_meta else None, uses_meta=uses_meta, tta_transforms=4)
        except Exception:
            img_tensor = infer_transform(img_pil).unsqueeze(0).to(DEVICE)
            with torch.no_grad():
                try:
                    out = model(img_tensor, meta_tensor) if uses_meta else model(img_tensor)
                except TypeError:
                    out = model(img_tensor)
                if isinstance(out, (tuple, list)): out = out[0]
                probs = torch.softmax(out, dim=1).cpu().numpy()[0]
        idx = int(np.argmax(probs)); conf = float(probs[idx])
        rows.append({"Model": name, "Prediction": CLASS_NAMES[idx], "Confidence": conf})
    df = pd.DataFrame(rows)

    fig1, ax1 = plt.subplots(figsize=(6,4))
    ax1.bar(df['Model'], df['Confidence']); ax1.set_ylim(0,1); ax1.set_ylabel("Confidence"); ax1.set_title("Model Confidence")
    plt.xticks(rotation=30); plt.tight_layout()

    ordered_names = list(model_files.keys()); n=len(ordered_names)
    fig2, axes = plt.subplots(1, max(1,n), figsize=(4*max(1,n),4))
    if n==1: axes=[axes]
    for ax, m_name in zip(axes, ordered_names):
        cm = confusion_mats.get(m_name)
        if cm is None:
            ax.text(0.5,0.5,"No CM saved", ha="center"); ax.axis('off')
        else:
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax, xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
            ax.set_title(m_name); ax.set_xlabel("Predicted"); ax.set_ylabel("True")
    plt.tight_layout()
    return df, fig1, fig2

with gr.Blocks() as demo:
    gr.Markdown("# Alzheimer's - Multi-Model Prediction (Image + Metadata)")
    gr.Markdown("Upload 2D MRI slice and provide Age, MMSE.")
    with gr.Row():
        with gr.Column(scale=1):
            img_in = gr.Image(type="pil", label="Upload MRI (2D slice)")
            age_in = gr.Number(value=float(meta_mean[0]), label="Age")
            mmse_in = gr.Number(value=float(meta_mean[1]), label="MMSE")
            btn = gr.Button("Run All Models")
        with gr.Column(scale=1):
            out_table = gr.Dataframe(headers=["Model","Prediction","Confidence"], label="Predictions")
            out_conf = gr.Plot(label="Confidence Comparison")
            out_cm = gr.Plot(label="Confusion Matrices (test set)")
    btn.click(fn=predict_all, inputs=[img_in, age_in, mmse_in], outputs=[out_table, out_conf, out_cm])

demo.launch(share=True)
```

```python
# ================== IMPORTS ==================
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt
from PIL import Image
import cv2

from captum.attr import LayerGradCam

# ================== CONFIG ==================
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
IMG_ROOT = "/content/drive/MyDrive/alzheimers-demo/extracted_images"
CLASS_NAMES = ["Healthy", "MCI", "Mild_AD"]
MODEL_PATH = "/content/drive/MyDrive/alzheimers-demo/models/resnet_showcase.pth"

# ================== DATA PIPELINE ==================
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

dataset = datasets.ImageFolder(root=IMG_ROOT, transform=train_transform)

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size
train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=16)
test_loader = DataLoader(test_ds, batch_size=1)

# ================== MODEL ==================
class ResNetSimple(nn.Module):
    def __init__(self, num_classes=3):
        super().__init__()
        self.backbone = models.resnet18(pretrained=True)
        in_features = self.backbone.fc.in_features
        self.backbone.fc = nn.Linear(in_features, num_classes)

    def forward(self, x):
        return self.backbone(x)

model = ResNetSimple(num_classes=3).to(DEVICE)

# ================== TRAIN (only 2 epochs) ==================
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

print("Training for 2 epochs just for explainability showcase...")
for epoch in range(2):
    model.train()
    total_loss = 0
    for imgs, labels in train_loader:
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}/2 - Loss: {total_loss/len(train_loader):.4f}")

torch.save(model.state_dict(), MODEL_PATH)
print(f"Model saved at {MODEL_PATH}")

# ================== GRAD-CAM FUNCTION ==================
def grad_cam_explanation(model, img_tensor, target_class):
    grad_cam = LayerGradCam(model.backbone, model.backbone.layer4)
    attribution = grad_cam.attribute(img_tensor.unsqueeze(0).to(DEVICE),
                                     target=target_class)
    attribution = attribution.squeeze().cpu().detach().numpy()

    attribution = np.maximum(attribution, 0)
    attribution = attribution / (attribution.max() + 1e-7)

    attribution_resized = cv2.resize(attribution, (224, 224), interpolation=cv2.INTER_CUBIC)
    return attribution_resized

# ================== PREDICTION + EXPLAIN ==================
def predict_and_explain(image_path):
    # Load image
    image = Image.open(image_path).convert("RGB")
    img_tensor = train_transform(image).to(DEVICE)

    # Prediction
    model.eval()
    with torch.no_grad():
        output = model(img_tensor.unsqueeze(0))
        probs = torch.softmax(output, dim=1)[0]
        pred_class = torch.argmax(probs).item()
        confidence = probs[pred_class].item()

    print(f"Prediction: {CLASS_NAMES[pred_class]} ({confidence*100:.2f}%)")

    # Grad-CAM
    grad_cam_attr = grad_cam_explanation(model, img_tensor, pred_class)

    # Overlay
    img_np = np.array(image.resize((224, 224))) / 255.0
    heatmap = cv2.applyColorMap(np.uint8(255 * grad_cam_attr), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB) / 255.0
    overlay = 0.5 * img_np + 0.5 * heatmap

    # Show
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.imshow(img_np)
    ax1.set_title("Original MRI")
    ax1.axis("off")

    ax2.imshow(overlay)
    ax2.set_title("Grad-CAM Overlay")
    ax2.axis("off")
    plt.show()

# ================== TEST WITH ONE IMAGE ==================
test_image_path = "/content/OAS1_0056_MR1_mpr-2_anon_sag_66.png"  # CHANGE this
predict_and_explain(test_image_path)
```
